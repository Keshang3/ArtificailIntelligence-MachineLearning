3 Image Compression and Decompression using PCA.
In this exercise, build a PCA from scratch using explained variance method for image compression task.
You are expected to compute the necessary matrices from the scratch. Dataset: Use image of your
choice.
1. Load and Prepare Data:
• Fetch an image of you choice.{If colour convert to grayscale}
• Center the dataset - Standaridze the Data.
• Calculate the covaraince matrix of the Standaridze data.

```python
mean_vec = np.mean(img_array, axis=0)
standardized_data = img_array - mean_vec
```

```python
# Convert image to 2D data
data = img_array.reshape(-1, img_array.shape[-1])

# Standardize
mean = np.mean(data, axis=0)
std = np.std(data, axis=0)

standardized_data = (data - mean) / std

# Number of samples
n = standardized_data.shape[0]

# Covariance matrix
covariance_matrix = (standardized_data.T @ standardized_data) / (n - 1)

print(covariance_matrix.shape)
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(img_array, cmap='gray')
plt.title("Original Grayscale")
```

```python
plt.subplot(1, 2, 2)
plt.imshow(covariance_matrix, cmap='hot')
plt.title("Covariance Matrix Visualization")
plt.colorbar()
plt.show()
```

```python
print(f"Standardized Data Shape: {standardized_data.shape}")
print(f"Covariance Matrix Shape: {covariance_matrix.shape}")
```

2. Eigen Decomposition and Identifying Principal Components:
• Compute Eigen Values and Eigen Vectors.
• Sort the eigenvalues in descending order and choose the top k eigenvectors corresponding to
the highest eigenvalues.
• Identify the Principal Components with the help of cumulative Sum plot.

```python
eigenvalues, eigenvectors = np.linalg.eigh(covariance_matrix)
```

```python
sorted_index = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_index]
sorted_eigenvectors = eigenvectors[:, sorted_index]
```

```python
total_variance = np.sum(sorted_eigenvalues)
explained_variance = sorted_eigenvalues / total_variance
cumulative_variance = np.cumsum(explained_variance)
```

```python
plt.figure(figsize=(10, 5))
plt.step(range(1, len(cumulative_variance) + 1), cumulative_variance, where='mid', label='Cumulative Explained Variance')
plt.axhline(y=0.95, color='r', linestyle='--', label='95% Threshold')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Ratio')
plt.title('Explained Variance vs. Number of Components')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

```python
k = np.argmax(cumulative_variance >= 0.95) + 1
print(f"To retain 95% of the image information, we need k = {k} components.")
```

3. Reconstruction and Experiment:

• Reconstruction: Transform the original data by multiplying it with the selected eigenvec-
tors(PCs) to obtain a lower-dimensional representation.

• Experiments: Pick Four different combination of principal components with various ex-
plained variance value and compare the result.

• Display the Results and Evaluate.

```python
def reconstruct_pca(standardized_data_pca, original_mean_flattened, original_std_flattened, eigenvectors, k, original_image_shape):
    selected_vectors = eigenvectors[:, :k]
    # Project data to lower dimension
    projected_data = np.dot(standardized_data_pca, selected_vectors)
    # Reconstruct data back to original feature space (still standardized)
    reconstructed_standardized_data = np.dot(projected_data, selected_vectors.T)
    # Inverse standardize: (standardized_data * std) + mean
    reconstructed_original_scale_data = (reconstructed_standardized_data * original_std_flattened) + original_mean_flattened
    # Clip values to be within [0, 255] for image display and convert to uint8
    reconstructed_original_scale_data = np.clip(reconstructed_original_scale_data, 0, 255).astype(np.uint8)
    # Reshape back to image dimensions
    reconstructed_img = reconstructed_original_scale_data.reshape(original_image_shape)
    return reconstructed_img
```

```python
# Define k values for experimentation based on explained variance
k_1 = 1 # Smallest k, high compression
k_lower_variance = np.argmax(cumulative_variance >= 0.85) + 1
k_mid_variance = np.argmax(cumulative_variance >= 0.90) + 1
k_high_variance = np.argmax(cumulative_variance >= 0.95) + 1 # This is the 'k' we found earlier

k_values = sorted(list(set([k_1, k_lower_variance, k_mid_variance, k_high_variance])))

plt.figure(figsize=(15, 10))

# Original image subplot
plt.subplot(2, 3, 1)
plt.imshow(img_array) # img_array is the original RGB image
plt.title("Original Image")
plt.axis('off')

# Reconstruct and display for each k value
for i, current_k in enumerate(k_values):
    # Pass the correct mean and std (from cell zPj2s-EgK4LB) and the original image shape
    reconstructed_img = reconstruct_pca(
        standardized_data,  # standardized_data from zPj2s-EgK4LB (pixels, channels)
        mean,               # mean from zPj2s-EgK4LB (channels,)
        std,                # std from zPj2s-EgK4LB (channels,)
        sorted_eigenvectors,
        current_k,
        img_array.shape     # original image shape (height, width, channels)
    )

    plt.subplot(2, 3, i + 2) # Start from second subplot
    plt.imshow(reconstructed_img)

    # Get explained variance for the current_k
    # Ensure current_k-1 is within bounds of cumulative_variance array
    if current_k - 1 < len(cumulative_variance):
        explained_var_at_k = cumulative_variance[current_k - 1] * 100
    else:
        # Fallback if k is larger than available components (shouldn't happen with argmax logic)
        explained_var_at_k = cumulative_variance[-1] * 100 # Take max variance if k is too large

    plt.title(f"k={current_k} ({explained_var_at_k:.2f}% var)")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
